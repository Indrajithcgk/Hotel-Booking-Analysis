# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Loading the Dataset

```python
df = pd.read_csv("hotel_bookings 2.csv")
```

# Exploratary Data Analysis and Data Cleaning

```python
print("Shape of dataframe :", df.shape)
print("\n","*"*50)
print(df.columns)
print("\n")
print(df.info())
```

```python
#Change reservation_status_date datatype to time data datatype
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

```python
df.describe(include = "object")
```

```python
#Get unique value of all categorical columns:
for col in df.describe(include = "object").columns:
    print(col)
    print(df[col].unique())
    print('-'*50)
```

```python
df.isnull().sum().sort_values(ascending = False)
```

```python
df.drop(["company", "agent"], axis = 1 , inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.describe()
```

```python
df= df[df["adr"]<5000]
```

```python
df.describe()
```

# Data  Analysis and Visualization

```python
cancelled_percentage = df["is_canceled"].value_counts(normalize=True)
print(cancelled_percentage)

plt.figure(figsize=(6,5))
plt.title("Reservation Status :")
plt.bar(["Not_cancelled", "Cancelled"], df["is_canceled"].value_counts(),ec =  "y",width = 0.7 )
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(x= "hotel", hue = "is_canceled",data  = df,palette = "Blues")
legend_labels = ax.get_legend_handles_labels()
plt.title("Reservation status in different hotels",size = 20)
plt.xlabel("Hotel")
plt.ylabel("Number of Reservations")
plt.show()
```

```python
resort_hotel = df[df["hotel"] == "Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
city_hotel = df[df["hotel"] == "City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel = city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title("Average Daily Rate(ADR) in City and Resort Hotel",fontsize= 30)
plt.plot(resort_hotel.index,resort_hotel["adr"],label = "resort_hotel")
plt.plot(city_hotel.index,city_hotel["adr"],label = "resort_hotel")
```

```python
df["month"] = df["reservation_status_date"].dt.month
```

```python
plt.figure(figsize=(16,8))
ax = sns.countplot(x= "month", hue = "is_canceled",data  = df,palette = "bright")
legend_labels = ax.get_legend_handles_labels()
plt.title("Reservation status in per Month",size = 20)
plt.xlabel("Month")
plt.ylabel("Number of Reservations")
plt.legend(["Not canceled", "canceled"])
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per Month", fontsize = 30)
sns.barplot('month', "adr", data = df[df["is_canceled"] == 1].groupby("month")[['adr']].sum().reset_index())
plt.show()
```

```python
canceled_data = df[df["is_canceled"] == 1]
top_10_country = canceled_data["country"].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title("Top 10 country with reservation canceled",fontsize = 15)
plt.pie(top_10_country,labels=top_10_country.index, autopct="%.2f")
plt.show()
```
